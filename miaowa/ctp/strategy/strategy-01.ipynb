```python
import ast
import re
```

```python
params = "['O', 'C', 'V', 'C.SMA(5)', 'C.SMA(10)', 'C.SMA(20)', 'C.RSI(6)', 'C.RSI(12)', 'C.MACD(61,7,12)', 'C.MACD(1,2,3)']"
param_list = ast.literal_eval(params)
```

```python
param_list
```

```python
for p in param_list:
#     pattern = re.compile(r"(?<=RSI\()\d+(?=\))")
    a = re.findall(r"(?<=RSI\()\d+(?=\))", p)
    print(a)
```

```python
periods = []
pattern = re.compile(r"(?<=MACD\()\d+,\d+,\d+(?=\))")
for param in param_list:
    p = pattern.findall(param)
    if p is not None and len(p) > 0:
        for n in p[0].split(','):
            periods.append(int(n))
        break
periods
```
